## Load Scores

```python
# Load Scores

score_dict = {'atrial': [], 'cfmature': [], 'cfwk8': [], 'epiicardium': [],
              'myocardium': [], 'smcwk19': [], 'smcwk8': [], 'valvelate': [], 'ventricular': [],
              'arteries': [], 'capillary': [], 'cfwk6': [], 'endocardium': [], 'lymphec': [],
              'neuralcrest': [], 'pericytes': [], 'smcwk6': [], 'valveearly': [], 'veins': [], 'all': []}

onehot = np.load('/mnt/lab_data3/soumyak/cardiogenesis/BPNET_deepshaps/onehot_peaks/all_onehot.npy')

for cluster in score_dict:
    
    print()
    print(cluster)
    
    ctype_dict = {i:[] for i in range(5)}
    for fold in range(5):
        print(fold)
        scores = np.load(pred_dir + '/' + cluster + '/' + 'fold' + str(fold) + '.preds.npy')
        print(scores.shape)
        ctype_dict[fold] = scores
    mean_score = np.mean(np.array([ctype_dict[i] for i in range(5)]), axis=0)
    print(mean_score.shape)
    
    with open(mean_dir + '/' + cluster + '.mean.preds.npy', 'wb') as outfile:
        np.save(outfile, mean_score)
        
#     with open(mean_dir + '/' + cluster + '.mean.count.shap.npy', 'rb') as infile:
#         score_dict[cluster] = np.load(infile)
```
