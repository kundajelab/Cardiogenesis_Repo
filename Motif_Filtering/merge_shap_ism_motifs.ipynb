## Imports

```python
import pandas as pd
import os
```

## Addresses

```python
shap_dir = '/oak/stanford/groups/akundaje/laks/illuminafiles/cardiomyogenesis/BPNET_deepshaps/shap_cleaned_motifs'
ism_dir = '/oak/stanford/groups/akundaje/laks/illuminafiles/cardiomyogenesis/BPNET_deepshaps/ism_cleaned_motifs'
all_dir = '/oak/stanford/groups/akundaje/laks/illuminafiles/cardiomyogenesis/BPNET_deepshaps/all_cleaned_motifs'
```

## Check Outdir

```python
if not os.path.isdir(all_dir):
    os.mkdir(all_dir)
```

## Load Shap Motifs

```python
shap_active_motifs = {'atrial': [], 'cfmature': [], 'cfwk8': [], 'epiicardium': [],
                      'myocardium': [], 'smcwk19': [], 'smcwk8': [], 'valvelate': [], 'ventricular': [],
                      'arteries': [], 'capillary': [], 'cfwk6': [], 'endocardium': [], 'lymphec': [],
                      'neuralcrest': [], 'pericytes': [], 'smcwk6': [], 'valveearly': [], 'veins': [], 'all': []}

for cluster in shap_active_motifs:
    print()
    print(cluster)
    shap_active_motifs[cluster] = pd.read_csv(shap_dir + '/' + cluster + '.active_motifs.bed', sep='\t', header=0)
    shap_active_motifs[cluster].columns = ['chr', 'start', 'end', 'tf_name', 'score', 'strand']
    
    display(shap_active_motifs[cluster].head())
    print()
    display(shap_active_motifs[cluster].shape)
```

## Load ISM Motifs

```python
ism_active_motifs = {'atrial': [], 'cfmature': [], 'cfwk8': [], 'epiicardium': [],
                     'myocardium': [], 'smcwk19': [], 'smcwk8': [], 'valvelate': [], 'ventricular': [],
                     'arteries': [], 'capillary': [], 'cfwk6': [], 'endocardium': [], 'lymphec': [],
                     'neuralcrest': [], 'pericytes': [], 'smcwk6': [], 'valveearly': [], 'veins': [], 'all': []}

for cluster in ism_active_motifs:
    print()
    print(cluster)
    ism_active_motifs[cluster] = pd.read_csv(ism_dir + '/' + cluster + '.active_motifs.bed', sep='\t', header=0)
    ism_active_motifs[cluster].columns = ['chr', 'start', 'end', 'tf_name', 'score', 'strand']
    
    display(ism_active_motifs[cluster].head())
    print()
    display(ism_active_motifs[cluster].shape)
```

## Merge Shap and ISM Motifs

```python
all_active_motifs = {'atrial': [], 'cfmature': [], 'cfwk8': [], 'epiicardium': [],
                     'myocardium': [], 'smcwk19': [], 'smcwk8': [], 'valvelate': [], 'ventricular': [],
                     'arteries': [], 'capillary': [], 'cfwk6': [], 'endocardium': [], 'lymphec': [],
                     'neuralcrest': [], 'pericytes': [], 'smcwk6': [], 'valveearly': [], 'veins': [], 'all': []}

for cluster in all_active_motifs:
    print()
    print(cluster)
    
    all_motifs = pd.concat([shap_active_motifs[cluster], ism_active_motifs[cluster]])
    print(all_motifs.shape)
    print()
    assert(all_motifs.shape[0] == (shap_active_motifs[cluster].shape[0] + ism_active_motifs[cluster].shape[0]))
    
    all_unique_motifs = all_motifs.drop_duplicates(subset=['chr', 'start', 'end', 'tf_name'])
    all_active_motifs[cluster] = all_unique_motifs
    
    display(all_active_motifs[cluster].head())
    print()
    display(all_active_motifs[cluster].shape)
```

## Save Merged Motifs

```python
for cluster in all_active_motifs:
    print(cluster)
    all_active_motifs[cluster].to_csv(all_dir + '/' + cluster + '.active_motifs.tsv', index=False, sep='\t')
```
